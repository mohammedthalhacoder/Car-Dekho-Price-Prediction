```python
# changing the columns name
data_1.rename(columns={'bt':'Body type', 'km':'Kilometers driven', 'transmission':'Transmission type', 'ownerNo':'Number of Owners', 'model': 'Car model', 'modelYear':'Year of manufacture', 'No Door Numbers':'No of doors'}, inplace=True)
```

```python
# changing the , and convert the dtype in Kilometers driven column

data_1['Kilometers driven'] = data_1['Kilometers driven'].str.replace(',', '').astype(int)
```

```python
# converting the lakh, crore, and thousand to numeric values

def convert_price(price_str):
    price_str = re.sub(r'[₹,]', '', price_str)

    if 'Lakh' in price_str:
        return float(price_str.replace('Lakh', '').strip()) * 10**5
    elif 'Crore' in price_str:
        return float(price_str.replace('Crore', '').strip()) * 10**7
    else:
        return float(price_str)
    
data_1['price'] = data_1['price'].apply(convert_price)
```

```python
data_1['Insurance Validity'].value_counts()
```

```python
# correcting the strings

data_1['Insurance Validity'] = data_1['Insurance Validity'].str.replace('Third Party insurance insurance', 'Third Party insurance')
data_1['Insurance Validity'] = data_1['Insurance Validity'].str.replace('1', 'First Party insurance')
data_1['Insurance Validity'] = data_1['Insurance Validity'].str.replace('2', 'Second Party insurance')
```

```python
# changing the dtype and removing the seats string in columns
data_1['Seats'] = data_1['Seats'].str.replace('Seats', '').astype(float)
```

```python
data_1.head()
```

```python
# changing the dtype and removing the string in Engine columns
data_1['Engine'] = data_1['Engine'].str.replace('CC', '').astype(float)
```

```python
# changing the dtype and removing the string in Mileage columns
data_1['Mileage'] = data_1['Mileage'].str.replace('kmpl', '').str.replace('km/kg','').astype(float)
```

```python
# changing the dtype and removing the string in Top Speed columns
data_1['Top Speed'] = data_1['Top Speed'].str.replace(r'[^\d.]', '', regex=True).replace('',np.nan)
data_1['Top Speed'] = data_1['Top Speed'].astype(float)
```

```python
# changing the dtype and removing the string in Cargo Volumn columns
data_1['Cargo Volumn'] = data_1['Cargo Volumn'].str.replace(r'[^\d.]', '', regex=True)
data_1['Cargo Volumn'] = data_1['Cargo Volumn'].astype(float)
```

```python
# replacing the correct string format in Steering Type column

replacement = {
    'power' : 'Power',
    'Electronic': 'Electric',
    'Electrical': 'Electric',
    'electric': 'Electric'
}


data_1['Steering Type'] = data_1['Steering Type'].replace(replacement)
```

```python
# standardize the formatting of text data
data_1['Turbo Charger'] = data_1['Turbo Charger'].str.capitalize()
```

```python
# standardize the formatting of text data
data_1['Super Charger'] = data_1['Super Charger'].str.capitalize()
```
